# Inverse encodin/ Inverse transform /Decoding categorical variables
Inverse encoding is the process of converting the encoded data back to its original form. This is done by using the inverse_transform() method. This method is available for the following encoders:
* LabelEncoder
* OrdinalEncoder
* OneHotEncoder

### Encoding
> ### Note: If we want to reverse the encoding, we have to create separate encoder for every single column
> This is because reverse encoder take a random column and select its data type and assign the values for all other columns.

```python
# generate the variables for every categorical variable using label encoding
le_sex = LabelEncoder()
le_embarked = LabelEncoder()
le_embark_town = LabelEncoder()
le_class = LabelEncoder()
le_who = LabelEncoder()
le_alive = LabelEncoder()

# now encoding
df['sex'] = le_sex.fit_transform(df['sex'])
df['embarked'] = le_embarked.fit_transform(df['embarked'])
df['embark_town'] = le_embark_town.fit_transform(df['embark_town'])
df['class'] = le_class.fit_transform(df['class'])
df['who'] = le_who.fit_transform(df['who'])
df['alive'] = le_alive.fit_transform(df['alive'])

# same work with loop
# columns_to_encode = ['sex', 'embarked', 'who', 'class', 'embark_town', 'alive']

# # Dictionary to store LabelEncoders for each column
# label_encoders = {}

# # Loop to apply LabelEncoder to each column
# for col in columns_to_encode:
#     # Create a new LabelEncoder for the column
#     le = LabelEncoder()

#     # Fit and transform the data, then inverse transform it
#     df[col] = le.fit_transform(df[col])

#     # Store the encoder in the dictionary
#     label_encoders[col] = le

# Check the first few rows of the DataFrame
df.head()
```

### Reverse Encoding

```python
# inverse the encoding
df['sex'] = le_sex.inverse_transform(df['sex'])
df['embarked'] = le_embarked.inverse_transform(df['embarked'])
df['embark_town'] = le_embark_town.inverse_transform(df['embark_town'])
df['class'] = le_class.inverse_transform(df['class'])
df['who'] = le_who.inverse_transform(df['who'])
df['alive'] = le_alive.inverse_transform(df['alive'])

df.head()
```

### OneHotEncoder

```python
# initialize the one hot encoder
ohe_sex = OneHotEncoder()
ohe_embarked = OneHotEncoder()
ohe_embark_town = OneHotEncoder()
ohe_class = OneHotEncoder()
ohe_who = OneHotEncoder()
ohe_alive = OneHotEncoder()

# now encoding
# use array at the end 
df['sex'] = ohe_sex.fit_transform(df[['sex']]).toarray()
df['embarked'] = ohe_embarked.fit_transform(df[['embarked']]).toarray()
df['embark_town'] = ohe_embark_town.fit_transform(df[['embark_town']]).toarray() 
df['class'] = ohe_class.fit_transform(df[['class']]).toarray()
df['who'] = ohe_who.fit_transform(df[['who']]).toarray()
df['alive'] = ohe_alive.fit_transform(df[['alive']]).toarray()

df.head()
```
