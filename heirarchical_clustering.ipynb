# Heirarchical Clustering
Heirachical clustering is a method for grouping data into disjoint subsets of similar data. It is a non-parametric method that is used when the number of clusters is not known in advance. It is also used to find clusters in data that is not normally distributed.

**Methods:**
* `Agglomerative Clustering` (Bottom up approach)
* `Divisive` (Top down approach)

**Methods to Find Similarities between Clusters:**
1. Min
2. Max

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
from sklearn.datasets import make_blobs

# remove warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Generate a random dataset with 100 samples and 3 features
X, y = make_blobs(n_samples=100, centers=11, n_features=3, random_state=42)

# Perform hierarchical clustering with the AgglomerativeClustering algorithm
clustering = AgglomerativeClustering(n_clusters=3, linkage='ward')
clustering.fit(X)

# Plot the dendrogram
plt.figure(figsize=(10, 7))
plt.title("Dendrogram")
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.show()
```

```python
# load tips dataset from sns
df = sns.load_dataset('tips')

# Generate a random dataset with 100 samples and 3 features
X = df[['total_bill', 'size', 'tip']]



# Perform hierarchical clustering with the AgglomerativeClustering algorithm
clustering = AgglomerativeClustering(n_clusters=3, linkage='ward')
clustering.fit(X)

# Plot the dendrogram
plt.figure(figsize=(10, 7))
plt.title("Dendrogram")
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.show()
```
